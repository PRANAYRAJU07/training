```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, ne_chunk
from nltk.probability import FreqDist
```

```python
text = "Artificial Intelligence is transforming the world. NLP is a key area of AI."

sentences = sent_tokenize(text)
print("Sentence Tokenization:")
print(sentences)

words = word_tokenize(text)
print("\nWord Tokenization:")
print(words)
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = [word for word in words if word.lower() not in stop_words and word.isalpha()]

print("After Stopword Removal:")
print(filtered_words)
```

```python
ps = PorterStemmer()

print("Stemming Results:")
for word in filtered_words:
    print(word, "→", ps.stem(word))
```

```python
lemmatizer = WordNetLemmatizer()

print("Lemmatization Results:")
for word in filtered_words:
    print(word, "→", lemmatizer.lemmatize(word))
```

```python
sentence = "Students are learning Natural Language Processing"

tokens = word_tokenize(sentence)
pos_tags = pos_tag(tokens)

print("POS Tagging:")
print(pos_tags)
```

```python
sentence = "Sundar Pichai is the CEO of Google and lives in California."

tokens = word_tokenize(sentence)
pos_tags = pos_tag(tokens)
ner_tree = ne_chunk(pos_tags)

print("Named Entity Recognition:")
print(ner_tree)
```

```python
text = "Data science uses data to extract meaningful insights from data"

tokens = word_tokenize(text.lower())
filtered_tokens = [word for word in tokens if word.isalpha()]

freq_dist = FreqDist(filtered_tokens)

print("Top 5 Most Frequent Words:")
print(freq_dist.most_common(5))
```
